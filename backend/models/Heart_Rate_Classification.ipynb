```python
def handle_outliers_iqr(df_cleaned, cols):
    for col in cols:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_cleaned[col] = np.clip(df_cleaned[col], lower, upper)
    return df_cleaned

cols_with_outliers = ['sdrr', 'pnn50', 'heart_rate', 'spo2', 'weekly_hours']
df_cleaned = handle_outliers_iqr(df_cleaned, cols_with_outliers)
```

```python
# Visualize outliers using box plots for numerical columns
numerical_cols = df_cleaned.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x=df_cleaned[col])
    plt.title(f'Box plot of {col}')
plt.tight_layout()
plt.show()
```

## cleaning data teks

```python
df_cleaned.head()
```

```python
for col in ['job_role']:
    df_cleaned[col] = df_cleaned[col].str.lower()

display(df_cleaned.head())
```

```python
import string
def remove_punctuatuion(text):
  if isinstance(text, str):
    return text.translate(str.maketrans('', '', string.punctuation))
  return text

for col in ['job_role']:
  df_cleaned[col] = df_cleaned[col].apply(remove_punctuatuion)

display(df_cleaned.head())
```

```python
# menghilangkan karakter nomor
import re
def remove_numbers(text):
    if isinstance(text, str):
        return re.sub(r'\d+', '', text)
    return text

for col in ['job_role']:
  df_cleaned[col] = df_cleaned[col].apply(remove_numbers)

display(df_cleaned.head())
```

```python
def remove_extra_spaces(text):
    if isinstance(text, str):
        return re.sub(r'\s+', ' ', text).strip()

    return text

for col in ['job_role']:
  df_cleaned[col] = df_cleaned[col].apply(remove_extra_spaces)

display(df_cleaned.head())
```

```python
def remove_non_ascii(text):
    if isinstance(text, str):
        return text.encode('ascii', 'ignore').decode('ascii')
    return text

for col in ['job_role']:
  df_cleaned[col] = df_cleaned[col].apply(remove_non_ascii)

display(df_cleaned.head())
```

```python
# Calculate correlation matrix for numerical features
numerical_df = df_cleaned.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()

print("Correlation Matrix for Numerical Features:")
display(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# One-hot encode fitur Job_Role
df_encoded = pd.get_dummies(df_cleaned, columns=['job_role'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler, PowerTransformer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

num_features = ['rmssd_log','heart_rate_log','spo2_log','pnn50_log','sdrr_log']
job_role_features = [col for col in df_encoded.columns if col.startswith('job_role')]
all_features = num_features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded[all_features])
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot PCA
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], alpha=0.6, color='skyblue')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA 2D of HRV Data')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoded['cluster'] = kmeans.fit_predict(X_pca)
```

```python
sil_score = silhouette_score(X_scaled, df_encoded['cluster'])
print("Silhouette Score:", sil_score)
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
cluster_gmm = gmm.fit_predict(X_pca)

sil = silhouette_score(X_pca, cluster_gmm)
print("Silhouette Score:", sil)
```

```python
df_encoded['gmm_cluster'] = cluster_gmm
df_encoded.groupby('gmm_cluster')[num_features].mean()
```

```python
cluster_means = df_encoded.groupby("gmm_cluster")[["rmssd_log"]].mean()
cluster_order = cluster_means["rmssd_log"].sort_values()

label_map = {
    cluster_order.index[0]: "medium",
    cluster_order.index[1]: "low",
    cluster_order.index[2]: "high"
}
```

```python
df_encoded["gmm_label"] = df_encoded["gmm_cluster"].map(label_map)
```

```python
import joblib
import pandas as pd

pipeline = joblib.load("gmm_pipeline.pkl")

# Ambil fitur yg dipakai saat training
feature_order = pipeline["scaler"].feature_names_in_

label_map = {0: "medium", 1: "low", 2: "high"}

def test_manual_input(pipeline):
    print("Masukkan data fisiologis untuk prediksi cluster:")

    # Input manual
    rmssd = float(input("RMSSD: "))1
    heart_rate = float(input("Heart Rate (bpm): "))
    spo2 = float(input("SPO2: "))
    pnn50 = float(input("pNN50: "))
    sdrr = float(input("SDRR: "))

    # Buat dict sesuai nama fitur training
    input_dict = {
      "rmssd_log": np.log(rmssd),
      "heart_rate_log": np.log(heart_rate),
      "spo2_log": np.log(spo2),
      "pnn50_log": np.log(pnn50),
      "sdrr_log": np.log(sdrr)
  }

    # SUSUN ulang kolom mengikuti training
    input_df = pd.DataFrame([[input_dict[f] for f in feature_order]], columns=feature_order)

    # Prediksi
    cluster = pipeline.predict(input_df)[0]
    label = label_map.get(cluster, "unknown")

    print("\n--- Hasil Prediksi ---")
    print("Cluster :", cluster)
    print("Label   :", label)


# Jalankan testing
test_manual_input(pipeline)
```

```python
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.utils.metric import distance_metric, type_metric

initial_medoids = np.random.choice(len(X_pca), 3, replace=False).tolist()

# metric
metric = distance_metric(type_metric.EUCLIDEAN)

# K-Medoids
kmed = kmedoids(X_pca, initial_medoids, metric=metric)
kmed.process()

clusters = kmed.get_clusters()
medoids = kmed.get_medoids()

print("Medoids index:", medoids)
print("Cluster sizes:", [len(c) for c in clusters])
```

```python
labels_kmed = np.zeros(len(X_pca), dtype=int)

for cluster_id, cluster_points in enumerate(clusters):
    for idx in cluster_points:
        labels_kmed[idx] = cluster_id

# Hitung silhouette
sil_kmed = silhouette_score(X_pca, labels_kmed)
print("Silhouette Score (K-Medoids):", sil_kmed)
```

```python
import hdbscan

clusterer = hdbscan.HDBSCAN(
    min_cluster_size=30,      # jumlah minimal anggota cluster
    min_samples=10,           # kontrol sensitivitas noise
    metric='euclidean',       # metric jarak
    cluster_selection_method='eom'   # hasil terbaik di banyak kasus
)

labels_hdb = clusterer.fit_predict(X_pca)
```

```python
unique, counts = np.unique(labels_hdb, return_counts=True)
list(zip(unique, counts))
```

```python
mask = labels_hdb != -1  # buang noise
if len(np.unique(labels_hdb[mask])) > 1:
    sil = silhouette_score(X_pca[mask], labels_hdb[mask])
    print("Silhouette Score (tanpa noise):", sil)
else:
    print("Silhouette Score tidak dapat dihitung karena hanya 1 cluster.")
```
